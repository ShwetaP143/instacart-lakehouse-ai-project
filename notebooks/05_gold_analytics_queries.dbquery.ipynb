```sql
-- 1. Total Orders
SELECT COUNT(*) AS total_orders
FROM main.instacart_gold.fact_order_items;

-- 2. Total Unique Customers
SELECT COUNT(DISTINCT user_id) AS total_customers
FROM main.instacart_silver.orders;

-- 3. Total Products Sold
SELECT COUNT(DISTINCT product_id) AS total_products_sold
FROM main.instacart_gold.fact_order_items;

-- 4. Top 5 Departments by Orders (Correct)
SELECT 
    dp.department,
    COUNT(*) AS total_orders
FROM main.instacart_gold.fact_order_items fo
JOIN main.instacart_gold.dim_products dp
ON fo.product_id = dp.product_id
GROUP BY dp.department
ORDER BY total_orders DESC
LIMIT 5;
```
